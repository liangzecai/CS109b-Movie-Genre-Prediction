# Part 3 Build From Scrath

## 3.1 Baseline MLP Model

#### Create data for MLP model:

```python
# read in each image, use following code
# read in sampled genre dataframe
genre = pd.read_csv('genre_sample.csv', index_col=0)
data_genre = np.array(genre)

# create a dataframe to store information of image
index = genre.index.values
data_image = []

for i in range(len(index)):
    x = np.load('npy/' + str(genre.index.values[i]) + '.npy')
    data_image.append(x.flatten())

data_image = np.array(data_image)
# center each poster by subtracting the mean of each row
data_image = data_image - np.mean(data_image, axis = 1).reshape(-1,1)

# for vgg16, use df_image.values[i][0].reshape(1,224,224,3) to reshape dataset
```

#### Sampled Data (n = 1000) for MLP model:
X: data_image (1000 x 150528)

y: data_genre (1000 x 8)

```python
from __future__ import print_function

import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import SGD
from keras import regularizers

import h5py

import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
def base_MLP(x, y, batch_size = 128, num_classes = 8, epochs = 100, unit = 20, lr = 0.1, momentum = 0.9,
             validation_ratio = 0.2, regularizer = None, verbose = 1):

    ### ------- Baseline MLP - Layers ------- ###
    model = Sequential()
    # add an input layer
    model.add(Dense(unit, activation='relu', input_shape=(150528,))) # input picture size: 224x224x3
    # this is our hidden layer
    model.add(Dense(unit, activation='relu', kernel_regularizer=None))
    # and an output layer
    model.add(Dense(num_classes, activation='sigmoid'))
    # prints out a summary of the model architecture
    #model.summary()

    ### ------- Baseline MLP - Optimization ------- ###
    sgd = SGD(lr=lr, momentum=momentum)  # momentum = [0.9, 0.99]
    model.compile(loss='binary_crossentropy',
                  optimizer=sgd,
                  metrics=['accuracy'])

    ### ------- Baseline MLP - Train Model ------- ###
    history = model.fit(x, y,
                        batch_size=batch_size,
                        epochs=epochs,
                        verbose=verbose,
                        validation_split=validation_ratio)
    
    return (model, history)
```

### Model 0: baseline MLP

Baseline 3-layer MLP Model:
- dense(ReLU)
- dense(ReLU)
- dense(sigmoid)

Baseline Parameters:
- batch_size = 128
- num_classes = 8 
- epochs = 100
- unit = 20 (# of nodes each layer)
- lr = 0.1 
- momentum = 0.9
- validation_ratio = 0.2 
- regularizer = None

```python
###### Baseline MLP - Parameters ######
model0, history0 = base_MLP(x = data_image, y = data_genre, batch_size = 128, num_classes = 8, 
                   epochs = 100, unit = 20, lr = 0.1, momentum = 0.9, 
             validation_ratio = 0.2, regularizer = None, verbose = 0) # verbose = 0 for no logging to stdout  
```

```python
### Baseline MLP - Evaluation ###
plt.plot(history0.history['acc'], color = 'blue', label = 'train')
plt.plot(history0.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 0')
plt.show()
```

The optimizer didn't converge to a good local minimum, as the accuracy for training data failed to increase after a small amount of epochs.

## 3.2 Tune MLP Model

### Model 1: smaller learning rate
- set lr = 0.01

```python
###### MLP Model1 - small learning rate ######
# lr = 0.01
model1, history1 = base_MLP(x = data_image, y = data_genre, batch_size = 128, num_classes = 8, 
                   epochs = 100, unit = 20, lr = 0.01, momentum = 0.9, 
             validation_ratio = 0.2, regularizer = None, verbose = 0) # verbose = 0 for no logging to stdout 
```

```python
### MLP Model1 - Evaluation ###
plt.plot(history1.history['acc'], color = 'blue', label = 'train')
plt.plot(history1.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 1')
plt.show()
```

### Model 2: add L2 regularization

- set regularizer = regularizers.l2(0.01)

```python
###### MLP Model2 - L2 Regularization ######
model2, history2 = base_MLP(x = data_image, y = data_genre, regularizer = regularizers.l2(0.01), verbose = 0) 
```

```python
### MLP Model2 - Evaluation ###
plt.plot(history2.history['acc'], color = 'blue', label = 'train')
plt.plot(history2.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 2')
plt.show()
```

### Model 3: smaller batch size

- set batch_size = 32
- keep regularizer = regularizers.l2(0.01)

```python
###### MLP Model3 - smaller batch_size ######
model3, history3 = base_MLP(x = data_image, y = data_genre, batch_size = 32,
                            regularizer = regularizers.l2(0.01), verbose = 0) 
```

```python
### MLP Model3 - Evaluation ###
plt.plot(history3.history['acc'], color = 'blue', label = 'train')
plt.plot(history3.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 3')
plt.show()
```

### Model 4: more nodes for each layer

- set unit = 40
- keep regularizer = regularizers.l2(0.01)

```python
###### MLP Model4 - More Nodes for Each Layer ######
model4, history4 = base_MLP(x = data_image, y = data_genre, unit = 40,
                            regularizer = regularizers.l2(0.01), verbose = 0) 
```

```python
### MLP Model4 - Evaluation ###
plt.plot(history4.history['acc'], color = 'blue', label = 'train')
plt.plot(history4.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 4')
plt.show()
```

### Model 5: more epochs

- set epochs = 200
- keep regularizer = regularizers.l2(0.01)

```python
###### MLP Model5 - more epochs ######
model5, history5 = base_MLP(x = data_image, y = data_genre, epochs = 200,
                            regularizer = regularizers.l2(0.01), verbose = 0) 
```

```python
### MLP Model5 - Evaluation ###
plt.plot(history5.history['acc'], color = 'blue', label = 'train')
plt.plot(history5.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 5')
plt.show()
```

### Model 6: more layers
- add one extra hidden layer
- add one dropout layer
- keep regularizer = regularizers.l2(0.01)
- slightly change the nodes for each layer
- reduce the batch_size a little to 64

```python
###### MLP Model6 - extra hidden layer and Dropout layer ######

from keras.layers import Dropout

model6 = Sequential()
model6.add(Dense(64, activation='relu', input_shape=(150528,))) # input picture size: 224x224x3
model6.add(Dense(32, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model6.add(Dropout(0.5))
model6.add(Dense(16, activation='relu', kernel_regularizer=regularizers.l2(0.01)))
model6.add(Dense(8, activation='sigmoid'))


sgd = SGD(lr=0.1, momentum=0.9)  # momentum = [0.9, 0.99]
model6.compile(loss='binary_crossentropy',
                  optimizer=sgd,
                  metrics=['accuracy'])


history6 = model6.fit(data_image, data_genre,
                    batch_size=64,
                    epochs=100,
                    verbose=0,
                    validation_split=0.2)
```

```python
### MLP Model6 - Evaluation ###
plt.plot(history6.history['acc'], color = 'blue', label = 'train')
plt.plot(history6.history['val_acc'], color = 'red', label = 'test')
plt.xlabel("epoch")
plt.ylabel("accuracy")
plt.legend()
plt.title('MLP - Model 6')
plt.show()
```

#### Discussion:
- For the baseline 3-layer MLP model trained with 1000 samples, we get a accuracy of around 0.5.
- Add a l2 regularizer (lambda = 0.01) increases the accuracy a bit.
- Smaller learning rate, smaller batch_size, bigger epochs, or more nodes for each layer doesn't increase the accuracy. Some actually make worse models.
- In general, the baseline model is not very optimal according to its accuracy plot. The accuracy rises within few epochs and then start to plateau. What we ideally want is a gradually increasing accuracy plot.
- With deeper layers (and dropout layer), the accuracy finally seems to be able to get out of the previous local minimum, and starts to increase.

So in our tuning process , the best contribution to accuracy comes from building more complex layers with regularization (dropout and l2).
