# Downloading the top 500 movies on TMBd from the IMDb database

```python
import numpy as np 
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import random

%matplotlib inline
```

```python
from imdb import IMDb
ia = IMDb()
```

```python
# import TMDb top 500 movies imdb_ids
id_list = pd.read_csv('popular_imdb_id.txt', header=None)
ID = np.array(id_list)
```

```python
# columns of interest 
columns = ['title', 'genres', 'director', 'year', 'rating', 'votes', 'runtimes', 
          'language codes', 'language', 'producer', 
         'mpaa', 'writer', 'top 250 rank', 'kind', 'country codes', 'countries',
         'cover url', 'aspect_ratio', 'production companies',
          'cinematographer'
         'plot outline', 'plot', 'cast', 'animation department', 'original music',
         'canonical title', 'editorial department', 'canonical title', 'long imdb title',
          'long imdb canonical title', 'smart canonical title', 
          'smart long imdb canonical title', 'full-size cover url']
index = range(1, len(ID)) 
df = pd.DataFrame(index = index, columns=columns)
df = df.fillna(0)
```

```python
for i in range(0, len(index)):
    movie = ia.get_movie(ID[i])
    ia.update(movie)
    keys = movie.keys()
    
    for j in range(0, len(columns)):
        if columns[j] in keys:
            if type(movie[columns[j]]) == list:
                result = str(movie[columns[j]])
            else:
                result = movie[columns[j]]
            df.iloc[i,j] = result
        else:
            df.iloc[i,j] = 'nan'
```
